# Q11. Get average Salary grouped by City and store in variable called avg_salary.

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'City': ['Dehli', 'Mumbai', 'Chennai', 'Kolkata', 'Banglore'],
    'salary': [55000, 72000, 48000, 80000, 61000]
})

avg_salary = df.groupby('City', as_index=False)['salary'].mean()
print(avg_salary)
```

# Q12. Create another DataFrame dept_df and merge with df horizontally. Store it in a variable called merged_df

dept_df columns:

| Name    | Department |
| ------- | ---------- |
| Alice   | HR         |
| Bob     | IT         |
| Charlie | Finance    |
| David   | Marketing  |
| Eva     | Admin      |




```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 22, 35, 28],
    'City': ['Dehli', 'Mumbai', 'Chennai', 'Kolkata', 'Banglore'],
    'Salary': [55000, 72000, 48000, 80000, 61000]
    })
dept_df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Admin']
    })
merged_df = pd.merge(df, dept_df, on='Name')
print(merged_df)
```

# Q13. Concatenate two DataFrames vertically.

```python
import pandas as pd
df1 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 22],
    'City': ['Dehli', 'Mumabi', 'Chennai']
})

df2 = pd.DataFrame({
    'Name': ['David', 'Eva'],
    'Age': [35, 28],
    'City': ['Kolkata', 'Banglore']
})

merged_df = pd.concat([df1, df2], axis=0)

print(merged_df)
```
